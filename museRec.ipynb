```python
import pandas as pd
```

```python
s =pd.read_csv("top_10000_1960-now.csv")
```

```python
pd.set_option('display.max_columns',None)
```

```python
s.head()
```

```python
s = s[['Track Name', 'Artist Name(s)', 'Album Name', 'Album Image URL', 'Track Preview URL', 'Artist Genres', 'Label']]
```

```python
s['Artist Name(s)1'] = s['Artist Name(s)'].str.replace(' ','')
```

```python
s['Album Name'] = s['Album Name'].str.replace(' ','')
s['Artist Genres'] = s['Artist Genres'].apply(lambda x: str(x).split(','))
```

```python
s['Label'] = s['Label'].str.replace(' ','')
```

```python
s['Artist Name(s)1'] = s['Artist Name(s)'].apply(lambda x: [x])
s['Album Name'] = s['Album Name'].apply(lambda x: [x])
s['Label'] = s['Label'].apply(lambda x: [x])
```

```python
s.head()
```

```python
s['tags'] = s['Artist Name(s)1']+s['Album Name']+s['Artist Genres']+s['Label']
```

```python
s.head()
```

```python
s= s[['Track Name','Album Image URL','Track Preview URL','tags','Artist Name(s)']]
```

```python
s.isna().sum()
```

```python
s.drop('Track Preview URL',axis=1,inplace=True)
```

```python
s.isna().sum()
```

```python
s = s.dropna()
```

```python
s.isna().sum()
```

```python
s['tags'] = s['tags'].apply(lambda x:[str(i) for i in x])
s['tags'] = s['tags'].apply(lambda x: ' '.join(x))
```

```python
s['tags'] = s['tags'].apply(lambda x: x.lower())
```

```python
s.head()
```

```python
s.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=9995, stop_words='english')
```

```python
vectors = cv.fit_transform(s['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vectors)
```

```python
sim[0]
```

```python
s[s['Artist Name(s)'] == 'Michael Jackson']
```

```python
s[s['Artist Name(s)'] == 'ABBA']
```

```python
def recommend(song):
    song_id = s[s['Track Name']==song].index[0]
    distances = sim[song_id]
    song_list = []
    song_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:11]
    for i in song_list:
        print(s.iloc[i[0]]['Track Name'])
```

```python
recommend('Following the Sun')
```

```python
recommend('Without Me')
```

```python
s.to_csv('final__songs.csv')
```
